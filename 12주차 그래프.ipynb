```python
# 문제: 최소 비용 도로망 구축하기
#. 그래프 응용: Kruskal 이용하여 MST 구한 다음 최소비용을 출력한다. 

from operator import itemgetter
class Graph():
    def __init__(self, size):
        self.SIZE = size
        self.graph = [ [0 for _ in range(size)] for _ in range(size)]
        
    def add_edge(self, fromVertex, endVertex, weight): # 간선 추가
        self.graph[fromVertex][endVertex] = weight # edge = 가중치 연결, 인접

    def print_graph(self, graph): # 주석이 없는 그래프 출력
        count = 0
        print(' ', end = ' ')
        
        for row in range(self.SIZE):
            for col in range(self.SIZE):
                print("%2d" % self.graph[row][col], end=' \t') 
                if self.graph[row][col] == self.graph[col][row]:
                    count += 0.5 * self.graph[row][col]
            print()
            
        print()
        print("cost: ", count)
        
    
    def dfs(self, current, visited_ary):
        visited_ary.append(current)
        for vertex in range(self.SIZE):
            if self.graph[current][vertex] != 0 and vertex not in visited_ary:
                self.dfs(vertex, visited_ary)
            
            
    def find_vertex(self, find_vtx):
        visited_ary = []
        self.dfs(0, visited_ary)
        
        if find_vtx in visited_ary:
            return True
        else:
            return False
    
    
    def create_sorted_no_duplicated_edges(self): # 가중치 인접행렬 무방향 그래프의(간선, 출발정점, 도착정점) 전처리
        edge_ary = [] 
        for i in range(self.SIZE):
            for k in range(self.SIZE):
                if self.graph[i][k] != 0:
                    edge_ary.append([self.graph[i][k], i, k])
                    
        edge_ary = sorted(edge_ary, key = itemgetter(0), reverse = True) # 내림차순 정렬: 최소비용만 유지
        new_ary = [edge_ary[i] for i in range(0, len(edge_ary), 2)]
        
        return new_ary
    
    
    def kruskal_with_min_spanning_tree(self, edges):
        # 크루스컬 알고리즘: edges 배열은 내림차순으로 정렬된 중복이 없는 (간선, 출발정점, 도착정점)들로 구성
        index = 0
        while len(edges) > self.SIZE - 1: # 간선의 개수가 '정점 개수-1'일 때까지 반복
            start = edges[index][1]
            end = edges[index][2]
            save_cost = edges[index][0]
            
            self.graph[start][end] = 0 # 간선 제거
            self.graph[end][start] = 0
            
            start_yn = self.find_vertex(start) # 그래프의 연결 여부 확인
            end_yn = self.find_vertex(end)
            
            if start_yn and end_yn:
                del edges[index] # 연결된 것은 삭제
            else: # 연결
                self.graph[start][end] = save_cost
                self.graph[end][start] = save_cost
                index += 1
                
#######################################
# 전역 변수 선언 #
G1  = None
name_ary = ['춘천', '서울', '속초', '대전', '광주', '부산']
춘천, 서울 , 속초, 대전, 광주, 부산 = 0,1,2,3,4,5

# 메인 코드 부분 #
if __name__ == "__main__":
    # 간선 정보 추가
    gSize = 6
    G1 = Graph(gSize)
    G1.add_edge(춘천, 서울, 10)
    G1.add_edge(춘천, 서울, 15)
    G1.add_edge(춘천, 서울, 10)
    G1.add_edge(춘천, 서울, 40)
    G1.add_edge(속초, 광주, 50)
    G1.add_edge(속초, 춘천, 15)
    G1.add_edge(속초, 서울, 40)
    G1.add_edge(대전, 서울, 12)
    G1.add_edge(대전, 광주, 20)
    G1.add_edge(대전, 부산, 30)
    G1.add_edge(광주, 서울, 50)
    G1.add_edge(광주, 대전, 20)
    G1.add_edge(광주, 부산, 25)
    G1.add_edge(부산, 대전, 30)
    G1.add_edge(부산, 광주, 25)
   
    
    print('## 자전거 도로 건설을 위한 전체 연결도 ##')
    G1.print_graph(G1)
    
    #kruskal 알고리즘 적용
    # 가중치 간선 목록 생성
    edge_ary = G1.create_sorted_no_duplicated_edges()
    # 최소 비용의 자전거 도로 연결
    G1.kruskal_with_min_spanning_tree(edge_ary)
    print('## 최소 비용의 자전거 도로 연결도 ##')
    G1.print_graph(G1)
            
            
    
```
