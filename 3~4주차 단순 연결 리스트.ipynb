# Singly linked list 구현

```python
####### 클래스 선언 부분 #######
# 노드 정의
class Node():
    def __init__ (self):
        self.data = None
        self.link = None

####### 함수 선언 부분 #######
# 단순 연결 리스트 출력
def print_Nodes(start):
    current  = start
    while current:
        print(current.data, end = '')
        current = current.link
        
    print("\n================")
    
    
# 1. 단순 연결 리스트 생성
def makeSimpleLinkedList(dataArray):
    global memory, head
    node = Node()
    node.data = dataArray[0]
    head = node
    memory.append(node)
    
    for data in dataArray[1:]:
        pre = node
        
        node = Node()
        node.data = data
        pre.link = node
        
        memory.append(node)
        
# 2. 단순 연결 리스트의 특정 노드 검색
def search_Node(findData):
    global head
    
    current = head
    while current:
        if current.data == findData:
            print("Found")
            return current
        current = current.link
    return Node() # 못찾으면

# 3. 단순 연결 리스트의 노드 삽입
def insert_Node(findData, insertData):
    global memory, head
    
    if head.data == findData:
        node = Node()
        node.data = insertData
        node.link = head
        head = node
        return
    
    current = head
    while current:
        pre = current
        currenr = current.link
        if current.data == findData:
            node = Node()
            node.data = insertData
            node.link = current
            pre.link = node
            return
        
    node = Node() # 마지막 노드로 삽입 및 연결
    node.data = insertData
    current.link = node
    
# 4. 단순 연결 리스트의 노드 삭제
def delete_Node(deleteData):
    global memory, head
    
    if head.data == deleteData: # 첫 번째 노드 삭제
        current = head # 삭제 노드
        head = head.link #링크 수정
        del(current) # 데이터 삭제
        print("delete!")
        return
    
    current = head
    while current:
        pre = current
        current = current.link
        if current.data == deleteData:
            pre.link = current.link
            del(current)
            print("delete!")
            return

## 전역 변수 선언 ##
memory = []
head= None
dataArray = ["A", "B", "C", "D", "E"]

# 메인 코드 실행
if __name__ == "__main__":
    # 1. 메모리에 연결리스트 노드 저장
    makeSimpleLinkedList(dataArray)
    print_Nodes(head)
    
    # 2. 연결리스트의 특정 노드 탐색
    fNode = search_Node("D")
    print(fNode.data)
    print("\n================")
    
    # 3. 연결리스트에 추가로 노드 삽입
    insert_Node("A", "->")
    print_Nodes(head)
    
    # 4. 기존 연결리스트의 특정 노드 삭제
    delete_Node("C")
    print_Nodes(head)
```

# 문제: 로또 추첨(p.156)
1~45 숫자 6개를 뽑는 로또 추첨 프로그램
뽑은 숫자는 올림차순대로 단순 연결 리스트에 저장

```python
import random

class Node():
    def __init__(self):
        self.data = None
        self.link = None
        
def print_Nodes(start):
    current  = start
    while current:
        print(current.data, end = ' ')
        current = current.link
        
def makeLottoLinkedList(num):
    global memory, head
    node = Node()
    node.data = num
    memory.append(node)
    
    if head == None: # 첫 번째 노드
        head = node
        return
    
    if head.data > num: # 첫 번째 노드보다 작을 때
        node.link = head
        head = node
        return
    current = head
    
    # 중간 노드로 삽입하는 경우
    while current.link != None:
        pre = current
        current = current.link
        if current.data > num:
            pre.link = node
            node.link = current
            return
    # 삽입하는 노드가 가장 큰 경우
    current.link = node
    
## 전역 변수 선언 ##
memory = []
head= None

# 메인 코드 실행
if __name__ == "__main__":
    lottoCount = 0
    
    while True:
        lotto = random.randint(1, 45)
        lottoCount += 1
        makeLottoLinkedList(lotto)
        if lottoCount >= 6:
            break
    print_Nodes(head)
    
```
