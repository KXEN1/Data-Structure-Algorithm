```python
# Bellmanford :음의 가중치 그래프

class BellmanFord:
    def __init__(self, graph, start):
        self.graph = graph
        self.start = start
        self.vertices = set()
        self.distance = {}
        self.predecessor = {}

    def initialize(self):
        for vertex in self.graph:
            self.vertices.add(vertex)
            self.distance[vertex] = float('infinity')
            self.predecessor[vertex] = None
            
        self.distance[self.start] = 0

    def relax(self, u, v, weight):
        if self.distance[u] + weight < self.distance[v]:
            self.distance[v] = self.distance[u] + weight
            self.predecessor[v] = u
    
    def bellman_ford(self):
        self.initialize()

        # relaxation을 |V|-1 수행
        for _ in range(len(self.vertices) - 1):
            for u in self.graph:
                for v, weight in self.graph[u].items():
                    self.relax(u, v, weight)
                
        # 음의 사이클 검사
        for u in self.graph:
            for v, weight in self.graph[u].items():
                if self.distance[u] + weight < self.distance[v]:
                    return "음의 사이클이 존재합니다."
                                          
        return self.distance, self.predecessor
            
# 예제 그래프의 인접 리스트: 강의 교안의 예시 그래프
graph = {
    's': {'a': 10, 'b': 5, 'd': 7}, 
    'a': {'b': 2, 'c': 1}, 
    'b': {'a': 3, 'c': 9, 'd': 2},
    'c': {'a': 4, 'b': 3},
    'd': {'s': 6}
}

# 시작 노드를 's'로 설정
start_node = 's'

# 벨만-포드 알고리즘 적용
bellman_ford = BellmanFord(graph, start_node)
result = bellman_ford.bellman_ford()

for row in result:
    print(row)
```
