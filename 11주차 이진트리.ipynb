```python
# 이진트리 순회: code8-2.py

class TreeNode():
    def __init__ (self):
        self.left = None
        self.data = None
        self.right = None
        
def preorder(node): # 전위 순회
    if node == None: # base case
        return
    print(node.data, end = '->')
    preorder(node.left)
    preorder(node.right)
    
def inorder(node): # 중위 순회
    if node == None:
        return
    inorder(node.left)
    print(node.data, end = '->')
    inorder(node.right)
    
def postorder(node): # 후위 순회
    if node == None:
        return
    postorder(node.left)
    postorder(node.right)
    print(node.data, end = '->')
    
# 이진 트리 생성
root = TreeNode() # 루트 노드
root.data = '화사' # 노드 객체에 데이터 저장

# 자식 노드 객체 생성
node1 = TreeNode()
node1.data = '솔라'
root.left = node1 # 부모 노드와 자식 노드 연결

node2 = TreeNode()
node2.data = '문별'
root.right = node2

node3 = TreeNode()
node3.data = '휘인'
node1.left = node3

# 트리 순회
print("전위 순회: ", end = "")
preorder(root)
print("None")
print("중위 순회: ", end = "")
inorder(root)
print("None")
print("후위 순회: ", end = "")
postorder(root)
print("None")
```

```python
# code8-4.py
# 이진탐색 트리에서 데이터 검색
# 검색할 이름을 input()함수로 입력 받아 검색하기
#1. 이진탐색트리 생성-삽입 연산
#2. 트리 순회
#3. 탐색

class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None
        
def preorder(node): # 전위 순회
    if node == None: # base case
        return
    print(node.data, end = '->')
    preorder(node.left)
    preorder(node.right)
    
def make_bst(name_ary):
    global root, memory
    # 루트 노드 생성 및 트리에 삽입
    node = TreeNode()
    node.data = name_ary[0]
    root = node
    memory.append(root)
    
    # 루트 노드의 자식 노드 생성 및 트리에 삽입
    for name in name_ary[1:]:
        node = TreeNode()
        node.data = name
        # 삽입연산
        current = root
        while True:
            if name < current.data: # 왼쪽
                if current.left is None: # current.data == None
                    current.left = node
                    break
                current = current.left
            else:
                if current.right is None:
                    current.right = node
                    break
                current = current.right
            
        memory.append(node)
    return memory
    
## 전역 변수 선언 부분 ##
memory = [] # 이진탐색트리의 노드 저장
root = None
nameAry = ['블랙핑크', '레드벨벳', '마마무', '에이핑크', '걸스데이', '트와이스']

# 이진탐색트리 생성
memory = make_bst(nameAry)

# 이진탐색트리 출력: 전위 순회 방식으로 출력
preorder(memory[0])
```
