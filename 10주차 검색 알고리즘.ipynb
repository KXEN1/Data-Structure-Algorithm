```python
# 순차 탐색: O(n)
# 이진 탐색: O(logn)
import random

## 함수 선언 부분 ##
def binSearch(ary, fData): # 이진 탐색
    pos = -1
    start = 0
    end = len(ary) - 1
    
    while(start <= end):
        mid = (start + end) // 2
        if fData == ary[mid]:
            pos = mid
            return pos
        elif fData > ary[mid]:
            start = mid + 1
        else:
            end = mid - 1
            
    return pos


def seqSearch(ary, fData): # 순차 탐색
    pos = -1
    size = len(ary)
    
    
    for i in range(size):
        
        if ary[i] == fData:
            pos = i
            break
        
    return pos

## 전역 변수 선언 부분 ##
random.seed(1)
dataAry = [random.randint(1, 100) for _ in range(20)]
sortedAry = sorted(dataAry) # 정렬: 이진 검색

## 메인 코드 부분 ##
if __name__ == "__main__":
    findData = int(input('* 찾을 값을 입력하세요. --> '))
    print('배열 --> ', dataAry)
    print('배열 --> ', sortedAry)
    
    #position = seqSearch(dataAry, findData)
    position = binSearch(sortedAry, findData)
    
    if position == -1:
        print(findData, '(이)가 없네요')
    else:
        print('비교한 데이터', findData, '(은)는 ', position, '위치에 있음')
```

```python
#Ex 13-01.py 교재 응용예제 01 편의점에서 판매된 물건 목록과 개수 세기
import random

def binSearch(ary, fData):
    pos = -1
    start = 0;
    end = len(ary)-1
    
    while (start <= end):
        mid = (start + end) // 2
        if fData == ary[mid]:
            pos = mid
            return pos
        elif fData > ary[mid]:
            start = mid + 1
        else:
            end = mid -1
        
    return pos

## 전역 변수 선언 부분 ##
dataAry = ['바나나맛우유', '레쓰비캔커피', '츄파츕스', '도시락', '삼다수', '코카콜라', '삼각김밥']
sellAry = [random.choice(dataAry) for _ in range(10)]

## 메인 코드 부분 ##
print("# 오늘 판매된 전체 물건(중복 O, 정렬 X)-->", sellAry)
sellAry.sort()
print("# 오늘 판매된 전체 물건(중복 O, 정렬 O)-->", sellAry)
sellProduct = list(set(sellAry))
print("# 오늘 판매된 전체 물건(중복 X)-->", sellProduct)


countList = [] #tuple 형태 데이터를 저장하는 배열: (아이템, 판매개수)
for product in sellProduct:
    count = 0
    pos = 0
    while(pos != -1):
        pos = binSearch(sellAry, product) # 이진검색을 활용하여 아이템의 위치를 찾기
        if pos != -1:
            count += 1 # 찾으면 1만큼 증가하기
            del(sellAry[pos]) # 카운트한 아이템은 삭제하기 (중복 배제)
    countList.append((product, count)) # 튜플 저장하기
    
print()
print("결산 결과 ==>", countList)
```

# #. 이진검색을 이용한 단어사전에서 단어찾기

```python
# 이진 검색 알고리즘을 사용하여 단어를 검색하기
# 단어가 정렬된 상태로 유지
"""
1. 입력 문장을 정렬된 단어 리스트로 변환
2. 이진 검색을 사용하여 특정 단어를 검색
3. 정렬된 단어 리스트와 검색 결과가 출력
4. 단어 빈도수 계산
"""
def create_sortrd_word_list(sentence):
    # 문장을 공백을 기준으로 단어(토큰)로 분리한 후 정렬하여 리스트에 저장.
    word_list = sentence.replace(".", "").split()
    word_list.sort()
    return word_list

def binSearch(ary, fData):
    pos = -1
    start = 0;
    end = len(ary)-1
    
    while (start <= end):
        mid = (start + end) // 2
        if fData == ary[mid]:
            pos = mid
            return pos
        elif fData > ary[mid]:
            start = mid + 1
        else:
            end = mid -1
        
    return pos

## 전역변수: 입력 문장 ##
input_sentence = "This is a sample sentence. This sentence contains some sample words."

## 정렬된 단어 리스트 생성 및 출력 ##
sorted_word_list = create_sortrd_word_list(input_sentence) # 영어 사전

## 검색 결과 출력 ##
print("정렬된 닽어 리스트:")
for word in sorted_word_list:
    print(word)
    
## 검색할 단어 받기 ##
target_word = input("검색할 단어==>")

## 이진 검색을 사용하여 단어 검색 ##
pos = binSearch(sorted_word_list, target_word)
print()
print(f"FOUND! {sorted_word_list[pos]}")

## 단어 빈도수 계산 ##
freq_word_list = []

for word in set(sorted_word_list):
    count = 0
    pos = 0
    while (pos != -1):
        pos = binSearch(sorted_word_list, word)
        if pos != -1:
            count += 1
            del(sorted_word_list[pos])
    freq_word_list.append((word, count))
    
print("단어 빈도수")
for word, count in freq_word_list:
    print(f"({word}, {count})")
```
