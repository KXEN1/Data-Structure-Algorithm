# Circlular linked list 구현(p.170~)

```python
# 1. 노드 객체 선언
class Node:
    def __init__(self):
        self.data = None
        self.link = None
        
# 2. 함수 선언
# 원형 연결 리스트 출력
def print_Nodes(start):
    current = start
    
    while current.link != start:
        print(current.data, end = "->") # => None 대신 start로 수정: 원형 연결 리스트
        current = current.link
        
    print(current.link.data) # 마지막 노드에 연결된 첫 번째
    print("\n=================")
    
# 원형 연결 리스트 생성
def makeCircularLinkedList(dataArray):
    global memory, head
    
    #first node를 헤드 노드에 연결
    node = Node()
    node.data = dataArray[0]
    head = node
    node.link = head # 원형 연결
    memory.append(node)
    
    # 새 node를 이전 노드(첫 번째 노드)와 연결하고 head 노드에 연결
    for data in dataArray[1:]:
        pre = node
        node = Node()
        node.data = data
        pre.link = node
        node.link = head # 원형 연결
        memory.append(node)
        
# 원형 리스트의 특정 노드 검색(p.187 5-1)
def search_Node(findData):
    current = head # 첫 번째 노드
    if current.data == findData:
        print("첫 번째 노드에서 찾음!")
        return current
    
    while current.link != head:
        current = current.link
        if current.data == findData:
            print("중간 노드에서 찾음!")
            return current
        
    print("찾는 노드 없음")
    return Node() # 못찾으면

# 원형 연결 리스트에 노드 삽입
def insert_Node(findData, insertData):
    global memory, head
    
    if head.data == findData:
        node = Node()
        node.data = insertData
        node.link = head
        #### last node 찾아서 head에 연결####
        last = head # 마지막 노드를 첫 번째 노드로 우선 지정
        while last.link != head: # 마지막 노드를 찾으면 반복 종료
            last = last.link # last를 다음 노드로 변경
            
        last.link = node # 마지막 노드의 링크에 새 노드 지정
        head = node # 첫 번째 노드로 지정
        return
        
        current = head # 두 번재 노드부터 중간노드로 삽입
        while current.link != head:
            pre = current
            current = current.link
            if current.data == findData:
                node = Node()
                node.data = insertData
                node.link = current
                pre.link = node
                return
        
        node = Node()
        node.data = insertData
        current.link = node
        node.link = head # 원형 처리
        
# 원형 연결 리스트에 노드 삭제
def delete_Node(deleteData):
    global memory, head
    
    if head.data == deleteData: # 첫 번째 노드 삭제
        current = head
        head = head.link # 링크 수정
        #### last node 찾아서 head에 연결 ####
        last = head
        while last.link != current: # 마지막 노드를 찾으면 반복 종료
            last = last.link # last를 다음 노드로 변경
        last.link = head # 마지막 노드의 링크에 head가 가리키는 노드 지정
        del(current) # 데이터 삭제
        print("delete!")
        return
    
    current = head
    while current.link != head:
        pre = current
        current = current.link
        if current.data == deleteData:
            pre.link = current.link
            del(current)
            print("delete!")
            return
        
####### 전역 변수 선언 #######
memory = []
head = None
dataArray = ["a", "b", "d", "g", "h"]

if __name__ == "__main__":
    makeCircularLinkedList(dataArray)
    print_Nodes(head)
    fNode = search_Node("d")
    print(fNode.data)
    
    insert_Node("a", "e")
    print_Nodes(head)
    insert_Node("b", "f")
    print_Nodes(head)
    insert_Node("d", "ff")
    print_Nodes(head)
    
    delete_Node("h")
    print_Nodes(head)
```
